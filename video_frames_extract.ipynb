```python
import cv2
import os
```

```python
# video_path = str(input("Enter the path to the video (including the name of the video). Press Enter to use the default path:\n") or "./GoodGateTestLensCorrect.mp4")
video_path = "./GoodGateTestLensCorrect.mp4"
vidcap = cv2.VideoCapture(video_path)
success, image = vidcap.read()
# count_initial = int(input("Enter starting frame of video capture, or press enter to use default value of 750:\n") or 750)
frame_count = 0
img_count = 0
if not os.path.isdir("./source/master"):
    os.makedirs("./source/master")
    print("Created ./source/master directory")
    os.system('echo "Created ./source/master directory"')
else:
    os.rename('./source/master/', './source/master_OLD')
    os.makedirs('./source/master')
    print('Directory master already exists. Renamed it to OLD_master to avoid data loss')
    os.system("echo 'Directory master already exists. Renamed it to OLD_master to avoid data loss'")
```

```python
def reach_init_val(count, init_val):
    while count < init_val:
        success, image = vidcap.read()
        count += 1
    return count
```

```python
# Set of frames where you can see the gate [750, 1980]
#1st half of split of interval
frame_count = reach_init_val(frame_count, 750)
while success and frame_count <= 1200:
    cv2.imwrite("./source/master/frame%s.jpg" % format(img_count, "04"), image)
    success, image = vidcap.read()
    print("Read frame #{}: {}".format(frame_count, success))
    os.system('echo "Read frame #{}: {}"'.format(frame_count, success))
    frame_count += 1
    img_count += 1
```
